```python
# Function that can return all rows of a specified value
def getrows(column_name, value):
    return df[df[column_name] == value]


# For difference between Male and Female
a = getrows('gender','Male')
b = getrows('gender','Female')
```

```python
# Boxplot for outlier detection

numerical_cols = ['avg_work_hours_per_day','avg_rest_hours_per_day', 'avg_sleep_hours_per_day',
                 'avg_exercise_hours_per_day', 'age_at_death']

plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(data=df, y=col, color='skyblue')
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```
